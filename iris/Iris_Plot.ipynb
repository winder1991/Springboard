```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
data = pd.read_excel('Iris.xlsx', index_col=0)
```

```python
print(data.shape)
```

```python
print(data.columns)
```

```python
data['Species'].value_counts()
```

```python
data.plot(kind='scatter', x= 'Sepal length', y = 'Sepal width')
plt.show()
```

```python
fig, ax = plt.subplots()

number_of_colors = 3
# a way to randomly specify color
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(number_of_colors)]

items= list(set(data['Species']))
c={}
for item in items:
    c[item]="#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
#c={items[0]: "blue", items[1]:"red", items[2]:"green"}
# draw points in different color
for item in c:
    x= list(data['Sepal length'][data['Species']==item])
    y= list(data['Sepal width'][data['Species']==item])
    ax.scatter(x, y, color=c[item], label=item)

ax.legend()
plt.show()
```
